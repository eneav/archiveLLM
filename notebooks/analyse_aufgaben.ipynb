```python
import pandas as pd     #für tabellen bzw datenverarbeitung
import matplotlib.pyplot as plt         #für diagramme 
import seaborn as sns
import json
import os
```

```python
import os
print(os.getcwd())
```

```python
#lädt alle json files und fügt sie zu einer Liste zusammen
#am ende in tablle umwandeln

json_paths = [
    "../data/WB1_Klausur_Variante_1.json",
    "../data/WB1_Klausur_Variante_2.json",
    "../data/WB1_Klausur_Variante_3.json",
    "../data/WB1_Klausur_Variante_4.json",
    "../data/WB1_Klausur_Variante_5.json",
    "../data/WB1_Klausur_Variante_6.json",
    "../data/WB1_Klausur_Variante_7.json",
    "../data/WB1_Klausur_Variante_8.json",
    "../data/WB1_Klausur_Variante_9.json",
    "../data/WB1_Klausur_Variante_10.json",
]



all_data = []       
for path in json_paths:                                      #iteriert über alle json files (json_paths)
    with open(path, 'r', encoding='utf-8') as f:
        data = json.load(f)                              # inhalt der json files wird in data gespeichert
        all_data.extend(data)                        #daten zur gesamten liste all_data hinzufügen



df = pd.DataFrame(all_data)           #tabelle aus dict all_data erstellen
df.head()               #kontrolle, zeigt die ersten 5 zeilen des dataframes an
```

```python
# e(figsize=(10, 4))
sns.countplot(data=df, x="thema", order=df["thema"].value_counts().index, palette="Blues")
plt.title("Themenverteilung in 10 Klausuren")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



#erstellt balken diagramm mit der anzahl der fragen pro thema
```

```python
# verteilung der aufgabentypen
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="aufgabentyp", order=df["aufgabentyp"].value_counts().index, palette="Purples")
plt.title("Aufgabentypen in 10 Klausuren")
plt.tight_layout()
plt.show()


#erstellt balken diagramm mit der anzahl der fragen pro aufgabentyp
```

```python
#Tabelle exportieren in output ordner 
df.to_csv("../output/alle_fragen_gesamt.csv", index=False)
print("csv datei wurde gespeichert")
```
